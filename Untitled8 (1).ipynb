```python
import feedparser
import urllib.request as libreq
import pandas as pd
```

Данные про кофе

```python
# data collection

query = "coffee"
max_amount = 10
url = f"https://export.arxiv.org/api/query?search_query=all:{query}&start=0&max_results={max_amount}"
with libreq.urlopen(url) as url:
  r = url.read()
print(r)
docs = feedparser.parse(r)
```

```python
%time

pubs = []
for entry in docs["entries"]:
  data = {"title": entry['title'],
          "summary": entry['summary'],
          "authors": [author['name'] for author in entry['authors']],
    }
  pubs.append(data)
```

```python
df = pd.DataFrame(pubs)
df
```

### Загрузка модели для извлечения ключевых слов с HuggingFace


```python
%%time
# keywords extraction

from transformers import pipeline

pipe = pipeline("text2text-generation", model="ilsilfverskiold/tech-keywords-extractor")
```

### Извлечение ключевых слов по загаловку и анотации


```python
from tqdm.auto import tqdm
```

```python
%%time
tqdm.pandas()
df["title_summary"] = df["title"] + ". " + df["summary"]

df["keywords"] = df["title"].progress_apply(lambda x: pipe(x)[0]["generated_text"].split(", "))
```

```python
%%time
df["keywords_"] = df["summary"].apply(lambda x: pipe(x)[0]["generated_text"].split(", "))
```

```python
df["keywords"] = df["keywords"].apply(lambda x: ", ".join(x))
df
```

```python
df.to_csv("coffee_data_260.csv", index=False)
```

### Построения графа ключевых слов

### Загрузка библиотек и данных

```python
import pandas as pd
import networkx as nx
from itertools import combinations, chain
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from operator import itemgetter
import plotly.graph_objects as go
```

```python
df = pd.read_csv("coffee_data_260.csv")
df.sample(10)
```

```python
klist = [[word.strip() for word in keywords.split(",") if word]
         for keywords in df["keywords"].tolist()]
klist[:3]
```

```python
edges = list(chain(*[list(combinations(words, 2)) for words in klist]))
edges = [tuple(sorted(edge)) for edge in edges]
weighted_edges = [(edge[0], edge[1], {"weight": edges.count(edge)}) for edge in set(edges)]
len(edges)
len(weighted_edges)
```

```python
G = nx.Graph()
G.add_edges_from(edges)
nx.draw(G, with_labels=False, font_weight='bold', node_size=30)
```

```python
import networkx as nx
from itertools import combinations

G_kw = nx.Graph()

for kw_list in df["keywords"]:
    for w1, w2 in combinations(set(kw_list), 2):
        if G_kw.has_edge(w1, w2):
            G_kw[w1][w2]["weight"] += 1
        else:
            G_kw.add_edge(w1, w2, weight=1)
nx.draw(G_kw, with_labels=False, font_weight='bold', node_size=30)
```

```python
print(G.number_of_nodes())
print(list(G.nodes)[:5])
```

```python
communities = nx.community.louvain_communities(G, resolution=0.9)
modularity = nx.community.modularity(G, communities)
print(f"# of clusters: {len(communities)}, Modularity: {modularity:.4f}")
```

```python
comm_data = [{"n_of_nodes": len(comm), "nodes": comm} for comm in communities]
cdf = pd.DataFrame(comm_data)
cdf.describe()
```

```python
top3_comm = cdf.nlargest(3, "n_of_nodes")
top3_comm
```

```python
closeness = nx.closeness_centrality(G)

top_closeness = sorted(closeness.items(), key=lambda x: x[1], reverse=True)[:10]

print("Top 10 ключевых слов по closeness centrality:")
for word, score in top_closeness:
    print(f"{word}: {score:.4f}")
```

```python
G_pub = nx.Graph()

for i, row_i in df.iterrows():
    for j, row_j in df.iterrows():
        if i >= j:
            continue
        shared_keywords = set(row_i["keywords"]) & set(row_j["keywords"])
        if shared_keywords:
            G_pub.add_edge(i, j, weight=len(shared_keywords))
nx.draw(G_pub, with_labels=False, font_weight='bold', node_size=30)
```

```python
def find_pub_id_by_title(title_query):
    matches = df[df["title"].str.contains(title_query, case=False, na=False)]
    if matches.empty:
        print("Публикация не найдена.")
        return None
    return matches.index[0], matches.iloc[0]["title"]
```

```python
def find_similar_publications_by_title(title_query, top_n=5):
    result = find_pub_id_by_title(title_query)
    if result is None:
        return []

    pub_id, matched_title = result
    print(f"\n🔍 Найдена публикация:\n{matched_title}\nID: {pub_id}\n")

    if pub_id not in G_pub:
        print("У публикации нет связей в графе.")
        return []

    neighbors = sorted(G_pub[pub_id].items(), key=lambda x: x[1]["weight"], reverse=True)
    print(f"\n📚 Похожие публикации:")
    for idx, edge in neighbors[:top_n]:
        print(f"- {df.iloc[idx]['title']} (общих ключевых слов: {edge['weight']})")
```
